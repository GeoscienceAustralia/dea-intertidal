```python
# Generate vrts for the dea-interdidal products
# Currently writes vrts to current dir. Might want to change where the output is written to. 
import s3fs
#s3 path
product_dir ='dea-public-data-dev/derivative/dea_intertidal/sample_data/tiles/'

products =['elevation.tif', 'oa_hat.tif', 'oa_offset_hightide.tif', 
           'elevation_uncertainty.tif', 'oa_hot.tif', 'oa_offset_lowtide.tif', 'exposure.tif', 'oa_lat.tif', 
           'oa_spread.tif', 'extents.tif', 'oa_lot.tif']
# products =['elevation.tif']
time_ranges = [
    "2016_2018",
    "2017_2019",
    "2018_2020",
    "2019_2021",
    "2020_2022"
]

fs = s3fs.S3FileSystem(anon=True)
for time_range in time_ranges:
    for product in products:
        # get a list of the cogs(tifs) you want to include
        cogs = fs.glob(f'{product_dir}/**/*{time_range}*{product}')
        # print(cogs)
        if len(cogs)>0:
            files = ' '.join(cogs).replace('dea-public-data-dev/','/vsicurl/https://dea-public-data-dev.s3-ap-southeast-2.amazonaws.com/')
            vrt_name = f'DEV_{time_range}_{product[0:-4]}.vrt'
            !gdalbuildvrt -srcnodata -999 -vrtnodata -999 {vrt_name} {files}
            !gdalinfo -stats {vrt_name}
            with open(vrt_name, "r") as f:
                lines = f.readlines()
            with open(vrt_name, "w") as f:
                for line in lines:
                    if "OverviewList" not in line.strip("\n"):
                        f.write(line)
        else:
            print(f'No rasters found for {product} in {time_range}')
```

```python
#Separate the debug tiffs out from the dea-intertidal products
import glob
import os
import shutil
all_files = glob.glob(f'data/**/*debug*',recursive=True)
for file in all_files:
    if not os.path.exists('debug/' + os.path.dirname(file)):
        os.makedirs('debug/' +os.path.dirname(file))
    shutil.move(file, 'debug/' +file)
```
